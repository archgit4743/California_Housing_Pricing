# Splitting into test and train sets based on median income

The sample dataset 'test' and 'train' must be representative of the full dataset 'hs'. Since income is an important attributeto predict house prices, the stratified split is done based on income. 

```python
#converting the continuous variable 'median_income' to a categorical variable
hs['income_cat']=pd.cut(hs.median_income,bins=[0,1.5,3.0,4.5,6,np.inf],labels=[1,2,3,4,5])
hs['income_cat'].hist()
```

```python
from sklearn.model_selection import StratifiedShuffleSplit
split=StratifiedShuffleSplit(n_splits=1,test_size=0.2,random_state=42)
for train_index, test_index in split.split(hs,hs['income_cat']):
    strat_train_set = hs.loc[train_index]
    strat_test_set = hs.loc[test_index]
```

Dropping the income_cat attribute so that the data is back to original data


```python
for i in (strat_test_set,strat_train_set):
    i.drop('income_cat',axis=1,inplace=True)
```

# Treating the missing values

The column total_bedrooms has missing values, imputing the column with median.
But before that the column 'ocean_proximity' has to be dropped since the feature is object dtype.

```python
hs_median=hs.drop(['ocean_proximity'],axis=1)
from sklearn.impute import SimpleImputer
imputer=SimpleImputer(strategy='median')
imputer=imputer.fit(hs_median)
imputer.statistics_
```

```python
x=imputer.transform(hs_median)
hs_tr=pd.DataFrame(x,columns=hs_median.columns)
```
